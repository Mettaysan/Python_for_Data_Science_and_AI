# **Step 1: Import libraries**

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# **Step 2: Nikkei 225 Index monthly graph**

```python
# get the Nikkei monthly historical data
nikkei = '^N225'
nikkeiData = yf.Ticker(nikkei)
nikkeiDf = nikkeiData.history(period='1d', start='2020-1-1', end='2023-9-30')
monthly_nikkei_data = nikkeiDf.resample('M').mean()
monthly_nikkei_data
```

```python
# get the close historical data only
monthly_nikkei_data_close = monthly_nikkei_data[['Close']]
monthly_nikkei_data_close
```

```python
# show only four decimal places
pd.options.display.float_format ='{:,.4f}'.format
monthly_nikkei_data_close
```

```python
# time-series graph of Nikkei 225 Index
plt.figure(figsize=(14,7))
plt.plot(monthly_nikkei_data_close.index, monthly_nikkei_data_close['Close'])
plt.title('Nikkei 225 Index time-series graph', fontsize=16)
plt.xlabel('Date (Monthly)', fontsize=14)
plt.ylabel('Monthly Close (Index)', fontsize=14)
plt.grid(True)
plt.show()
```

*- 31st January - 31st March 2020
(2 months: drop)*

*- 1st April 2020 - 28th February 2021
(11 months: surge)*

*- 1st March 2021 - 31st January 2023
(23 months: stagnant)*

*- 1st February - 30th September 2023
(8 months: surge)*

# **Step 3: Nikkei 225 Index’s monthly rate of return time-series graph**

```python
# calculate the monthly rate of return
monthly_nikkei_data_close['Monthly rate of Return'] = monthly_nikkei_data_close['Close'].pct_change()
monthly_nikkei_data_close
```

```python
# remove NA values from 'monthly Return'
monthly_nikkei_data_return = monthly_nikkei_data_close['Monthly rate of Return'].dropna()
monthly_nikkei_data_return
```

```python
# plot the line graph of the rate of return
plt.figure(figsize=(14,7))
plt.plot(monthly_nikkei_data_return.index, monthly_nikkei_data_return)
plt.title('Monthly Rate of Return of Nikkei 225', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Monthly Rate of Return', fontsize=14)
plt.grid(True)
plt.show()
```

*- The lowest month was dropped to -18.15% (31st March 2020).*

*- The rate mostly wings around plus minus 5%.*

## **Q1**

*Q1. On average, did the Nikkei 225 Index yield a positive monthly rate of return during 31st January 2020 - 30th September 2023 (44 months)?*

## **A1**

```python
# calculate geometric mean of Nikkei monthly return
nikkei_geometric_monthly_return = np.prod(monthly_nikkei_data_return + 1)**(1/len(monthly_nikkei_data_return)) - 1
print(f"The geometric mean of monthly return for Nikkei 225 is {nikkei_geometric_monthly_return:.4f}")
```

*A1. On average, the Nikkei 225 Index yield a positive monthly rate of return of 0.74% during 31st January 2020 - 30th September 2023 (44 months).*

---



## **Q2**

*Q2. What was the Nikkei 225 Index Sharpe Ratio during 31st January 2020 - 30th September 2023 (44 months)?*

```python
# calculate standard deviation of Nikkei monthly return
nikkei_std_dev = monthly_nikkei_data_return.std()
print(f"The standard deviation of the monthly rate of return for Nikkei 225 is {nikkei_std_dev:.4f}")
```

```python
# Sharpe ratio of Nikkei 225 Index
risk_free_rate = -0.001/12
sharpe_ratio = (nikkei_geometric_monthly_return - risk_free_rate) / nikkei_std_dev
print(f"The Sharpe Ratio of the monthly rate of return for Nikkei 225 is {sharpe_ratio:.4f}")
```

## **A2**

*A2. Sharpe Ratio = 0.1605*
*- Return is 0.1605 units per unit of risk.*
