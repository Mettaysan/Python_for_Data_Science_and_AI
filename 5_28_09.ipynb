# **2. Work with Pandas**

```python
# check the population of Lao PDR, USA and Japan over time in dataframe
df = wb.data.DataFrame(
    'SP.POP.TOTL',
    ['LAO', 'USA', 'JPN'],
    db=57,
    version='202309',
    time=range(1970, 2023, 5),
)

df
```

```python
# check population and GDP of 3 Latin America countries
wb.data.DataFrame(['SP.POP.TOTL', 'NY.GDP.PCAP.CD'],
                  ['BRA', 'COL', 'ARG'],
                  db=57,
                  version='202309',
                  time=range(2010, 2015))
```

```python
# show dimensions for WDI archives
wb.source.concepts(db=1)
```

```python
# fetch GDP from an old version of the WDI
wb.data.DataFrame('NY.GDP.MKTP.CD',
                  ['CAN', 'FRA'],
                  version=201504,
                  db=57,
                  skipBlanks=True,
                  index='time')
```

```python
# population estimates for Brazil
df = wb.data.DataFrame(
    'SP.POP.TOTL',
    'BRA',
    db=57,
    time=range(1980, 2020),
    index='time',
    version=range(201004, 202104, 100)
)

df = df /1000000

df.plot(title='Brazil population estimates by WDI version').\
        set_ylabel('Millions')
```

```python
# check country code
wb.economy.coder(['Lao PDR', 'Japan'])
```
