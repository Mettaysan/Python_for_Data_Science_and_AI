# **3. Charting and plotting**

# **6. Aggregations**

```python
# let's check the Access to electricity (% of population) of Laos and Vietnam
df = wb.data.DataFrame(
    'EG.ELC.ACCS.ZS',
    ['LAO', 'VNM'],
    time=range(2000, 2023),
    #index='time'
    )

df
```

```python
# calculate unweighted averages for each year
df.mean()
```

```python
# calculate the weighted (respect to population) averages for each year
combined = wb.data.DataFrame(
    ['EG.ELC.ACCS.ZS', 'SP.POP.TOTL'],
    ['LAO', 'VNM'],
    time=range(2000, 2023),
    index=['series', 'economy']
)

combined
```

*Since there are a NaN values in the dataframe, we have to mask out such values.*

---



```python
# mask out the null values
elec_access = combined.xs('EG.ELC.ACCS.ZS')

combined = combined[elec_access.isna() == False]
combined
```

```python
# calculate the weighted average
pop = combined.xs('SP.POP.TOTL')
(elec_access*pop).sum() / pop.sum()
```

# **7. Group aggregations**

```python
# create a dataframe name c
c = wb.economy.DataFrame()
c
```

```python
# create a new column named Initial and take only the first alphabet of each row from column name
c['Initial'] = c['name'].str[0]
c
```

```python
# create a new dataframe that takes only the index column and Initial column
initial = c['Initial']
initial
```

```python
# create a dataframe that has both Access to electricity (% of population) and total population
combined = wb.data.DataFrame(
    ['EG.ELC.ACCS.ZS', 'SP.POP.TOTL'],
    skipAggs=True,
    time=range(2000, 2023),
    index=['series', 'economy']
)

combined
```

```python
# show the cross-section from the electricity
elec_access = combined.xs('EG.ELC.ACCS.ZS')
elec_access
```

```python
# mask out the null values
combined = combined[elec_access.isna() == False]
combined
```

```python
# get the cross-section of total population
pop = combined.xs('SP.POP.TOTL')
pop
```

```python
# calculate the weighted avarage
numerator = (elec_access * pop).join(initial).groupby('Initial').sum()
denominator = pop.join(initial).groupby('Initial').sum()

numerator / denominator
```
